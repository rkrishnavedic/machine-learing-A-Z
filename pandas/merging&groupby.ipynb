```python
d1 = pd.DataFrame({
    'city':['lucknow', 'kanpur', 'agra', 'delhi'],
    'temperature':[32,45,30,40]
})
```

```python
d1
```

```python
d2 = pd.DataFrame({
    'city':['kanpur','lucknow', 'delhi'],
    'humidity':[68,65,75]
})
```

```python
d2
```

```python
df = pd.merge(d1,d2, on='city')
```

```python
df
```

```python
pd.merge(d1,d2, on='city', how='outer')
```

```python
pd.merge(d1,d2, on='city', how='left')
```
