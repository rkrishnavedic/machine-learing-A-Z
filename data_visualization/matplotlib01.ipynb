```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
x = np.linspace(5,50,50)
y = np.sign(np.sin(x))
```

```python
plt.xlabel("current")
plt.ylabel("something")

plt.plot(x,y)

plt.xlim(30,40)
```

```python
x = np.linspace(2,10,10)
y = x

plt.plot(x,y,'g^', x,y**2, 'r-', x, y**3, '*')
```

```python
plt.figure(1)

plt.subplot(221)
plt.title("y=log(x)")
plt.plot(x,np.log(x))

plt.subplot(222)
plt.title("y=exp(x)")
plt.plot(x,np.exp(x))

plt.figure(2)

plt.subplot(2,2,1)
plt.title("y=x")
plt.plot(x,x)

plt.subplot(2,2,2)
plt.title("y=x^2")
plt.plot(x,x**2)

plt.show()
```
