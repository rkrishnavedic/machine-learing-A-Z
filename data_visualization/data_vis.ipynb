```python
df = pd.read_csv("./Datavisual/global_sales_data/market_fact.csv")
df = df[(df.Profit <100) & (df.Profit>-100) & (df.Sales<200)]

sns.jointplot('Sales', 'Profit', df, kind='hex', color='k')
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```
